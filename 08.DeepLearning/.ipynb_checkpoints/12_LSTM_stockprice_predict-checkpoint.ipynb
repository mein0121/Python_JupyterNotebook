##    
- X: 50 , y: 51 . 
    - 50   51  .
     

X, y .
 index window size   .

```python
#  input    
window_size = 50
data_X = [] # [[1-50],[2-51],[3-52],.....] 
data_y = [] # [51, 52, 53,.....]

for i in range(0,len(y)-window_size):
    data_X.append(X[i:i+window_size])
    data_y.append(y[i+window_size])
    
```

```python
np.shape(data_X)
```

```python
np.shape(data_y)
```

## Train, Test 

```python
# train:test 8:2
train_index = int(len(data_y)*0.8)
print(train_index)
```

```python
X_train, y_train = np.array(data_X[0:train_index]), np.array(data_y[0:train_index])
X_test, y_test = np.array(data_X[train_index:]), np.array(data_y[train_index:])
```

```python
X_train.shape, X_test.shape
```

## Model 
- LSTM  return_sequences               . 
    - return_sequences=True:   . False():  Sequence    
- many to many   LSTM     return_sequence=True  . 
- Dese False   .
-    return_sequences=False ,  return_sequence=True  .



<center>[: https://tykimos.github.io/2017/04/09/RNN_Getting_Started]</center>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
#  
LEARNING_RATE = 0.001
N_EPOCHS = 100
N_BATCHS = 100
N_TRAIN = X_train.shape[0]
N_TEST = X_test.shape[0]
```

```python
# Dataset 
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(N_TRAIN).batch(N_BATCHS, drop_remainder=True).repeat()
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(N_BATCHS)
```

```python
X_train.shape
```

```python
#  
def create_model():
    model = keras.Sequential()
    model.add(layers.Input((window_size,5)))
    # LSTM
    model.add(layers.LSTM(32, activation='relu', return_sequences=False))
    model.add(layers.Dense(32, activation='relu'))
    
    model.add(layers.Dense(1)) # (Regression), activation X
    
    return model
```

```python
model = create_model()
model.compile(optimizer = keras.optimizers.Adam(learning_rate=LEARNING_RATE),
              loss='mse')
model.summary()
```

```python
steps_per_epoch = N_TRAIN//N_BATCHS
validation_steps = int(np.ceil(N_TEST/N_BATCHS))
history = model.fit(train_dataset, epochs=N_EPOCHS, steps_per_epoch=steps_per_epoch, validation_data=test_dataset, validation_steps=validation_steps)
```

##  , 

```python
#
loss =model.evaluate(test_dataset)
```

```python
y_scaler.inverse_transform(np.array(loss).reshape(-1,1))#[np.newaxis,...])
```
